    - **popularity**:
   This column likely represents the popularity of a song, usually measured by metrics like number of streams, downloads, or views. It indicates how well-received or widely listened to a song is.

   - **acousticness**:
   This column measures the degree of acousticness of a song, indicating how much of the sound is derived from acoustic instruments as opposed to electronic ones. A higher value suggests a more acoustic sound.

   - **danceability**:
   This column measures how suitable a song is for dancing based on various musical elements such as rhythm, tempo, and beat strength. Higher values indicate songs that are easier to dance to.

   - **duration_ms**:
   This column represents the duration of the song in milliseconds, indicating how long the song lasts from start to finish.

   - **energy**:
   This column measures the energy level of a song. Songs with high energy are typically fast, loud, and noisy, while songs with low energy are often slow and calm.

   - **instrumentalness**:
   This column measures the presence of instrumental sounds in a song. A high instrumentalness value suggests that the song is predominantly instrumental, while a low value indicates the presence of vocals.

   - **liveness**:
   This column measures the presence of a live audience in the recording. A high value suggests that the song was performed live, while a low value indicates a studio recording.

   - **loudness**:
   This column represents the overall loudness of the song in decibels (dB). It indicates the volume level of the song.

   - **speechiness**:
   This column measures the presence of spoken words in the song, such as rap or spoken word segments. Higher values suggest a greater presence of speech-like elements.

   - **tempo**:
    This column represents the tempo of the song, measured in beats per minute (BPM). It indicates the speed or pace of the song.

   - **valence**:
    This column measures the overall positivity or happiness of the song. Higher values indicate a more positive or happy mood, while lower values suggest a more negative or sad mood.

   - **genre**:
    This column represents the genre or category of the song. It indicates the style or classification of music the song belongs to, such as pop, rock, jazz, etc.
